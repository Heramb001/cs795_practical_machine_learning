## Handling missing data
### pandas primarily uses the value np.nan to represent missing data. It is by default not included in computations.

### drap any rows with missing data

```python
df2 = df_train
df2.dropna(how = 'any')
```

### fill missing data with specific value

```python
df2 = df_train
df2.fillna(value = 5)
```

### get boolean mask where values are NaN

```python
df2 = df_train
pd.isna(df2)
```
